# Screenplay Genre Classification 

## Data Collection

```python
#importing necessary
import requests
from bs4 import BeautifulSoup
import os
import convertapi
```

### The titles and links to the screenplays I extracted were from the same site but from two different html links. The first html link is for movies A-M. The second html link is for movies N-Z. Both sets of screenplays are extracted from The Script Savant.

```python
#scrapping titles and pdf links from script-savant (movies only A to M)
req = requests.get("https://thescriptsavant.com/free-movie-screenplays-am/")
soup = BeautifulSoup(req.text, 'html.parser')

#initializing dictionary to store title and pdf link to movie
movie_dct = {}


for i in soup.find('tr').find_all('a'):
    if i.find('u') == None:
        continue
    else:
        movie_dct[i.find('u').text] = i['href'] #appends to dict
```

#### Keep in mind, the html format was different for both links. Therefore, the beautiful soup syntax is different for screenplays labelled N to Z.

```python
#scrapping titles and pdf links from script savant (movies only N to Z)
req = requests.get("https://thescriptsavant.com/free-movie-screenplays-nz/")
soup = BeautifulSoup(req.text, 'html.parser')

num = 0
for i in soup.find('div', class_='fusion-text').find_all('p'):
    if num <=2:
        num +=1 #skipping the first three rows because they were not pdf links to screenplays
    else:
        for x in i.find_all('a'):
            if x.find('u') == None: #checks if there is a pdf hyperlink (some screenplays didn't have hyperlinks)
                continue
            else:
                movie_dct[x.find('u').text] = x['href'] #adds to dict
```

```python
#downloading pdf files
for key, value in movie_dct.items():

    if os.path.exists("new_scripts/{}.pdf".format(key)) == True: #checks if screenplay exists in folder--this was used because there were failures in runtime
        continue                                                 #avoids double downloading
    else: 
        print(key)
        
        url = value
        r = requests.get(url, stream=True)

        try:
            with open("new_scripts/{}.pdf".format(key), "wb") as fd:
                for ch in r.iter_content():
                    fd.write(ch) #writes in the content from the hyperlink into a pdf file

            print('Success')
        except:
            pass


```

### There were numerous ways to convert pdf files to txt files but I ultimately decided to use the ConvertApi: https://www.convertapi.com/ . 

```python
#converting pdf files to txt files using convertapi

convertapi.api_secret = 'YOUR API SECRET' #you can get a free code on their website by signing up

for key, value in movie_dct.items():
    if os.path.exists("new_scripts/{}.pdf".format(key)) == True:  #checks if screenplay pdf exists
        if os.path.exists("script_texts/{}.txt".format(key)) == True: #checks if screenplay pdf was already converted -- this was used when computer failed during conversions
            continue                                                   #avoids double converting
        else:
            print(key)
            try:
                convertapi.convert('txt', {'File': 'new_scripts/{}.pdf'.format(key)}, from_format = 'pdf').save_files('script_texts/') #converts from pdf to txt
                print('Success')                                                                                                       #saves txt file to new folder
            except:
                pass
    else:
        continue
```
